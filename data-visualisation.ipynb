```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df.columns
```

```python
plt.scatter(df['age'], df['fare'])
plt.title('Visualisation')
plt.show()
```

```python
plt.bar(df['passenger_id'], df['fare'], color = 'red')
plt.show()
```

```python
plt.scatter(df['survived'], df['fare'], c = df['passenger_id'])
plt.title('Scatter with colors')
plt.show()
```

```python
plt.hist(df['fare'])
plt.title('Histogram of Fare')
plt.show()
```

```python
plt.boxplot('passenger_id', data=df)
plt.show()
```

```python
sns.scatterplot(x = 'passenger_id', y = 'fare', data = df)
plt.title('Scatter plot using Seaborn')
plt.show()
```

```python
sns.scatterplot(x = 'passenger_id', y = 'parch', data = df, hue = 'sex')
plt.title('Scatter plot using Seaborn')
plt.show()
```

```python
sns.lineplot(x = 'age', y = 'fare', data = df, hue= 'sex')
plt.title('Lineplot using Seaborn')
plt.show()
```

```python
sns.lineplot(data = df.drop(['passenger_id'], axis = 1))
plt.show()
```
